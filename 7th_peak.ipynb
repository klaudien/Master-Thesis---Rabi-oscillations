```python
data_23 = pd.read_csv('Scan23.csv')
cols_23 = list(data_23.columns)

plt.figure(figsize=(20,6))
plt.plot(data_23[cols_23[0]])
```

```python
def func(x,A,omega,phi,c,d,e):
    return A*np.sin(omega*x+phi)*np.exp(-c*x) - d*x + e
```

```python
x = [1e-5*x for x in range(0,len(data_22[cols_22[0]]))]

parameters = []

for num in range(0,1):
    
    para,pcov = curve_fit(func,x,data_22[cols_22[num]],p0=[0.025,1,np.pi/3,0.005,-0.041,0.574])

    y = [func(i,*para) for i in x]
    
    x = [1e-2*x for x in range(0,len(data_22[cols_22[0]]))]

    plt.figure(figsize=(20,6))
    plt.scatter(x,data_22[cols_22[num]],color='blue',label='Data')
    plt.plot(x,y,color='orange',label='Fit to the data with f='+str(round(para[1]/2/np.pi/1000,3))+'kHz and amplitude A='+str(round(para[0],3))+'V')
    plt.legend(fontsize=14)
    plt.title('Rabi oscillations for the 7th peak, time resolution=10$^{-5}$s',fontsize=14)
    plt.margins(x=0)
    plt.xlabel('Time, ms',fontsize=14)
    plt.ylabel('Signal, V',fontsize=14)
    
plt.savefig('7th_peak_comparison_1')
```

```python
x = [2e-5*x for x in range(0,len(data_23[cols_23[0]]))]

parameters = []

for num in range(0,1):
    
    para,pcov = curve_fit(func,x,data_23[cols_23[num]],p0=[0.0025,0.02,np.pi/3,0.05,-0.041,0.574])

    y = [func(i,*para) for i in x]

    x = [2e-2*x for x in range(0,len(data_23[cols_23[0]]))]
    
    plt.figure(figsize=(20,6))
    plt.scatter(x,data_23[cols_23[num]],color='blue',label='Data')
    plt.plot(x,y,color='orange',label='Fit to the data with f='+str((-1)*round(para[1]/2/np.pi/1000,3))+'kHz and amplitude A='+str(round(para[0],3))+'V')
    plt.legend(fontsize=14)
    plt.ylabel('Signal, V',fontsize=14)
    plt.xlabel('Time, ms',fontsize=14)
    plt.title('Rabi oscillations for the 7th peak, time resolution=2$\cdot$10$^{-5}$s',fontsize=14)
    plt.margins(x=0)
    
plt.savefig('7th_peak_comparison_2')
    
    
```

```python
data1 = np.array(pd.read_csv('Scan1(spec).csv'))

plt.figure(figsize=(20,7))

x= [x/100 for x in range(919000,919500,2)]

plt.plot(x[210:],data1[210:]*(-1)+0.3122)
plt.margins(x=0)
plt.ylabel('Signal, V')
plt.xlabel('Frequency, GHz')

plt.savefig('Spectro')
```

```python
def lorentzian( x, x0, a, gam ):
    return a * gam**2 / ( gam**2 + ( x - x0 )**2)
```

```python
data = np.ndarray.flatten(data1[210:]*(-1)+0.3122)
```

```python
para, pcov = curve_fit(lorentzian,x[210:],data,p0=[9194.38,0.024,0.05])
```

```python
plt.plot(x[210:],lorentzian(x[210:],*para),color='orange')
plt.scatter(x[210:],data1[210:]*(-1)+0.3122)
```
